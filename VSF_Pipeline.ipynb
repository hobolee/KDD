### Clone the repository

```python
%%sh
git clone https://github.com/google/vsf-time-series.git
```

### Install requirements

```python
%%sh
cd vsf-time-series/ && pip install -r requirements.txt
```

### Download Solar Data, Unzip it and Generate training dataset

#### Solar data

```python
%%sh
wget https://github.com/laiguokun/multivariate-time-series-data/raw/7f402f185cc2435b5e66aed13a3b560ed142e023/solar-energy/solar_AL.txt.gz -P vsf-time-series/data/ \
&& gunzip vsf-time-series/data/solar_AL.txt.gz \
&& mkdir -p vsf-time-series/data/SOLAR \
&& cd vsf-time-series \
&& python generate_training_data.py --ds_name solar --output_dir data/SOLAR --dataset_filename data/solar_AL.txt
```

#### Traffic data

```python
%%sh
wget https://github.com/laiguokun/multivariate-time-series-data/raw/7f402f185cc2435b5e66aed13a3b560ed142e023/traffic/traffic.txt.gz -P vsf-time-series/data/ \
&& gunzip vsf-time-series/data/traffic.txt.gz \
&& mkdir -p vsf-time-series/data/TRAFFIC \
&& cd vsf-time-series \
&& python generate_training_data.py --ds_name traffic --output_dir data/TRAFFIC --dataset_filename data/traffic.txt
```

### Training

```python
%%sh
cd vsf-time-series \
&& python train_multi_step.py --data ./data/TRAFFIC/ \
--model_name tri --device cuda:0 --expid 1 --epochs 1 --batch_size 64 --runs 10 \
--random_node_idx_split_runs 1 --lower_limit_random_node_selections 15 \
--upper_limit_random_node_selections 15 --step_size1 100 --mask_remaining true
```

```python
%cd /content
```

```python
%%sh
python train_multi_step.py --data ./data/Traffic --model_name tri \
--device cuda:0 --expid 1 --epochs 0 --batch_size 64 --runs 10 \
--random_node_idx_split_runs 100 --lower_limit_random_node_selections 15 \
--upper_limit_random_node_selections 15 --mask_remaining True
```
